```python
import pandas as pd
```

```python
recipes_no_image = pd.read_json('recipes_cleaned_1.json')
recipes_images = pd.read_json('recipes.json')
```

```python
#from recipes_images, we only need the image column and merge it with recipes_no_image on the name column
recipes_images = recipes_images[['name', 'img']]
recipes_no_image_new = pd.merge(recipes_no_image, recipes_images, on='name', how='left')
```

```python
#get number of duplicates in the name column
recipes_no_image_new['name'].duplicated().sum()
#remove duplicates
recipes_no_image_new = recipes_no_image_new.drop_duplicates(subset=['name'], keep='first')
```

```python
#save to json
recipes_no_image_new.to_json('recipes_cleaned_2.json', orient='records', indent=4, force_ascii=False)
```
